```python
t = 0
q = 0
prod = set()
for k in  dev_data.type_to_entityset["PROD"]:
    if k in uncased_wiki_data:
        #print(uncased_wiki_data[k])
        prod = prod.union(set(uncased_wiki_data[k]))
        t += 1
        if " " in k:
            q += 1
print(t/len(dev_data.type_to_entityset["PROD"]))
print(q/t)
print(prod)
        
```
